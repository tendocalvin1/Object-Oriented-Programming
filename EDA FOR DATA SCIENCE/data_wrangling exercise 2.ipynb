### Evaluate the dataset for missing data
Missing data was last converted from '?' to numpy.NaN. Pandas uses NaN and Null values interchangeably. This means, you can just identify the entries having Null values. Write a code that identifies which columns have missing data. 


````python
# Write your code below and press Shift+Enter to execute 
df.replace("?", np.nan, inplace=True) # common way to represent missing values in data analysis
missing_data = df.isnull() # This creates a new data frame(missing_data) with the same dimensions as df
# each cell in the missing_data data frame contains a boolean value: either True or False.[TRUE if the corresponding cell in df is missing]
# false if the cell in df has a value.
missing_data.head(5)

````

# Task - 2

### Replace with mean
Missing values in attributes that have continuous data are best replaced using Mean value. We note that values in "Weight_kg" attribute are continuous in nature, and some values are missing. Therefore, write a code to replace the missing values of weight with the average value of the attribute.


````python
# Write your code below and press Shift+Enter to execute
avg_weight = df["Weight_kg"].astype("float").mean(axis=0)
print("Average Weight in kg: ",avg_weight)
````

### Replace with the most frequent value
Missing values in attributes that have categorical data are best replaced using the most frequent value. We note that values in "Screen_Size_cm" attribute are categorical in nature, and some values are missing. Therefore, write a code to replace the missing values of Screen Size with the most frequent value of the attribute.


````python
# Write your code below and press Shift+Enter to execute
most_frequent_value = df["Screen_Size_cm"].mode()[0] # most frequent value.
# Replacing missing values with the most frequent value.[ We use the fillna() method ]
df["Screen_Size_cm"] = df["Screen_Size_cm"].fillna(most_frequent_value)
print(df)
````

Explanation:
 * Import Pandas: Imports the Pandas library, essential for data manipulation and analysis in Python.
 * Sample Data: Creates a sample DataFrame with a column named Screen_Size_cm containing categorical data, including some missing values (represented by None).
 * Find Most Frequent Value: Uses the mode() function to find the most frequent value in the Screen_Size_cm column. The [0] indexing is used to get the first mode (in case there are multiple modes).
 * Replace Missing Values: Uses the fillna() function to replace missing values (None) in the Screen_Size_cm column with the most_frequent_value calculated in the previous step.
 * Print DataFrame: Prints the modified DataFrame to display the results.

 Additional Considerations:
 * Handling Missing Values: While replacing with the most frequent value is a simple approach, consider other strategies like imputation using statistical methods (e.g., mean, median) or machine learning techniques based on the data and the specific use case.
 * Categorical Data: Ensure that the column you're working with is indeed categorical. If it's numerical, you might need different imputation techniques.
 * Data Quality: Before replacing missing values, it's good practice to check for data quality issues like outliers or inconsistencies that might affect the imputation results.
 * Domain Knowledge: Consider the domain knowledge and the impact of replacing missing values with the most frequent value. In some cases, it might be better to leave missing values as they are or use more sophisticated imputation methods.

### Fixing the data types
Both "Weight_kg" and "Screen_Size_cm" are seen to have the data type "Object", while both of them should be having a data type of "float". Write a code to fix the data type of these two columns.


````python
# Write your code below and press Shift+Enter to execute
df[["Weight_kg","Screen_Size_cm"]] = df[["Weight_kg","Screen_Size_cm"]].astype(object)
df.dtypes
````

This code snippet is designed to convert the data types of two specific columns in a Pandas DataFrame, Weight_kg and Screen_Size_cm, from their current data type (likely object or string) to the float data type.
Step-by-Step Explanation:

 * Converting Data Types:
   * .astype(float): This method is applied to the selected columns to convert their data type to float. This is crucial for numerical calculations and analysis.
Why Convert to float?

 Additional Considerations:

# Task - 4

### Data Standardization
The value of Screen_size usually has a standard unit of inches. Similarly, weight of the laptop is needed to be in pounds. Use the below mentioned units of conversion and write a code to modify the columns of the dataframe accordingly. Update their names as well.

```{math}
1 inch = 2.54 cm
1 kg   = 2.205 pounds
```


````python
# Write your code below and press Shift+Enter to execute
df[["Screen_Size_cm"]] = df[["Screen_Size_cm"]] / 2.54
df[["Weight_kg"]] = df[["Weight_kg"]] * 2.54

df.head()



````

### Data Normalization
Often it is required to normalize a continuous data attribute. Write a code to normalize the "CPU_frequency" attribute with respect to the maximum value available in the dataset.


````python
# Write your code below and press Shift+Enter to execute
df['CPU_frequency'] = df['CPU_frequency']/df['CPU_frequency'].max()
````

# Task - 5

### Binning
Binning is a process of creating a categorical attribute which splits the values of a continuous data into a specified number of groups. In this case, write a code to create 3 bins for the attribute "Price". These bins would be named "Low", "Medium" and "High". The new attribute will be named "Price-binned".


````python
# Write your code below and press Shift+Enter to execute
df["Price"]=df["Price"].astype(int, copy=True)
````

Also, plot the bar graph of these bins.


````python
# Write your code below and press Shift+Enter to execute
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["Price"])

# set x/y labels and plot title
plt.pyplot.xlabel("Price")
plt.pyplot.ylabel("count")
plt.pyplot.title("Price bins")
````
