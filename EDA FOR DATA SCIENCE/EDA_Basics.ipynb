# Duplicate data refers to records that have the same or nearly identical values across one or more key attributes. These duplicates can exist within a single dataset or across multiple datasets.

# Types of duplicates
1. Exact duplicates: These are the most straightforward, where all values in the record are identical.
2. Partial duplicates: These records have some matching values, but not all. For example, two records might have the same name and address, but have different phone numbers.
3. Near duplicates: These records have similar but not identical values. For example, names might be slightly misspelled or dates might be recorded with minor variations.

# Importance of identifying and handling duplicate data:
1. Data Quality: Duplicate data significantly impacts data quality. It introduces inconsistencies, skews statistical analysis, and leads to inaccurate insights.
2. Data Storage and Processing: Duplicates waste storage space and increase processing time during data cleaning, transformation, and analysis.
3. Machine Learning model Performance: Duplicates can negatively impact the performance of ML models.They can introduce bias, reduce model accuracy, and make it difficult to generalize from the data.
4. Decision Making: Incorrect or misleading information due to duplicates can lead to flawed business decisions, resulting in financial losses and reputational damage.
5. Data Integrity: Ensuring data integrity is crucial for maintaining trust and reliability in data-driven applications.


# Methods for identifying and handling duplicates
1.data profiling: Analyzing data characteristics to identify potential potential duplicates.
2. Deduplication Algorithms: Employing techniques like record linkage, clustering and fuzzy matching to identify and group similar records.
3. Data Cleaning Techniques: Removing or merging duplicate records based on specific rules and criteria.

# what is an outlier?
An outlier is a data point which is significantly far away from the other data points.
IQR(Inter Quartile Range)
IQR = Q3 - Q1
Lower Boundary = Q1 - (1.5 * IQR)
Upper Boundary = Q3 + (1.5 * IQR)
NB: When trying to detect outliers, we use a boxplot.
